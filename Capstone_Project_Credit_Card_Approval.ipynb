# Capstone Project Credit Card Approval

```python
import pandas as pd
```

### Reading two CSV files and merging them

```python
df1 = pd.read_csv('Credit_card.csv')
df2 = pd.read_csv('Credit_card_label.csv')

df = pd.merge(df1, df2, on='Ind_ID', how='inner')

df.head(10)
```

### Checking there are any NULL values or not

```python
df.info()
```

### Replacing all NaN values in GENDER column with the privious values

```python
df['GENDER'].fillna(method='ffill', inplace=True)  # Forward fill
df.info()
```

### Replacing all NaN values in Annual_income column with the privious values

```python
df['Annual_income'].fillna(method='ffill', inplace=True)  # Forward fill
df.info()
```

### Replacing all NaN values in Birthday_count column with the mean values

```python
df['Birthday_count'].fillna(df['Birthday_count'].mean(), inplace=True)
df.info()
```

```python
df.head(70)
```

### Deleting 'Type_Occupation' column because it is having more than 30% NaN values

```python
df1 = df.drop('Type_Occupation', axis=1)
df1
```

### checking the types of categorical veriables and Encoding them into numarical veriables, concat and drop the old column

```python
df4['GENDER'].value_counts()
```

```python
df_gen = pd.get_dummies(df4['GENDER'], drop_first = True)
df_gen1 = pd.concat([df4, df_gen], axis = 1)
df5 = df_gen1.drop('GENDER', axis=1)
df5
```

```python
df5['Car_Owner'].value_counts()
```

```python
df5['Propert_Owner'].value_counts()
```

```python
var_list = ['Car_Owner', 'Propert_Owner']

def binary_map(x):
    return x.map({'Y':1, 'N':0})

df5[var_list] = df5[var_list].apply(binary_map)
df5
```

```python
df5['Type_Income'].value_counts()
```

```python
df_it = pd.get_dummies(df5['Type_Income'], drop_first = True)
df_it1 = pd.concat([df5, df_it], axis = 1)
df6 = df_it1.drop('Type_Income', axis=1)
```

```python
df6['EDUCATION'].value_counts()
```

```python
df6['EDUCATION'] = df6['EDUCATION'].map({'Higher education':4, 'Incomplete higher':3, 'Secondary / secondary special':2, 'Lower secondary':1, 'Academic degree':0})
```

```python
df6['Marital_status'].value_counts()
```

```python
df_ms = pd.get_dummies(df6['Marital_status'], drop_first = True)
df_ms1 = pd.concat([df6, df_ms], axis = 1)
df7 = df_ms1.drop('Marital_status', axis=1)
```

```python
df7['Housing_type'].value_counts()
```

```python
df_ht = pd.get_dummies(df7['Housing_type'], drop_first = True)
df_ht1 = pd.concat([df7, df_ht], axis = 1)
df8 = df_ht1.drop('Housing_type', axis=1)
df8
```

### As the Data is imbalanced I am using Tree based models like Decision tree or Random forest

### Splitting the Features and Target 

```python
X = df8.drop(columns = 'label')
Y = df8['label']
```

```python
X.shape
```

```python
Y.shape
```

### Spliting the dataset into train set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, stratify=Y, random_state=25)
```

```python
#from sklearn.preprocessing import MinMaxScaler #Normalization
from sklearn.preprocessing import StandardScaler #Standardization
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Applying Machine Learning


#### As the output is in '0' and '1' means 'yes' and 'no' format, It is a Classification problem. 

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
Y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pred)
```

### We can see that after using logistic Regression we got 90% accuracy

### Tuning the ML model to improve the accuracy.
#### Hyper Parameter Tuning

```python
log_reg = LogisticRegression(penalty = 'l1', solver = 'liblinear')
log_reg.fit(X_train, Y_train)
```

```python
y_predt = log_reg.predict(X_test)
```

```python
accuracy_score(Y_test, y_predt)
```

```python
y_predt_train = log_reg.predict(X_train)
accuracy_score(Y_train, y_predt_train)
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier()
dec_tree.fit(X_train, Y_train)
```

```python
y_predd = dec_tree.predict(X_test)
accuracy_score(Y_test, y_predd)
```

```python
y_predd_train = dec_tree.predict(X_train)
accuracy_score(Y_train, y_predd_train)
```

### We can see Decision tree having overfitting issue

```python
dec_tree = DecisionTreeClassifier(max_depth = 15, max_features = 12)
dec_tree.fit(X_train, Y_train)
```

```python
y_predd = dec_tree.predict(X_test)
accuracy_score(Y_test, y_predd)
```

```python
y_predd_train = dec_tree.predict(X_train)
accuracy_score(Y_train, y_predd_train)
```
